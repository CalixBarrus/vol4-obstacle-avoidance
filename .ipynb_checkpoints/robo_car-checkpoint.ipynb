```python
import numpy as np
from scipy.integrate import odeint
from matplotlib import pyplot as plt
from scipy.linalg import solve_continuous_are
```

### Initialization

```python
V_nom = 10
d = 1 #Radius of the car
a = .001 # Delay constant for changing car's velocities
q_1 = 10000 # Penalty coefficient for being away from the wall
q_2 = 10 # Penalty coefficient for not being straight
q_3 = 1
q_4 = 1
r = .001 # Penalty coefficient for high command velocities to the wheels
initial_state = np.array([1, np.pi / 12, 0, 0])
```

```python
#LQR Matrices

A = np.array([[0, V_nom, 0, 0], [0, 0, -.5 * d, .5 * d], [0, 0, -1 * a, 0], [0, 0, 0, -1 * a]])
B = np.array([[0, 0], [0, 0], [a, 0], [0, a]])
Q = np.diag([q_1, q_2, q_3, q_4])
R = r * np.eye(2)
R_inverse = np.eye(2) / r
```

### Solve Infinite Horizon LQR

```python
P = solve_continuous_are(A, B, Q, R)
```

```python
def ode(z, t):
        #return A@z[:4] + B*z[4]
        return (A - B@R_inverse@B.T@P)@z

time_steps = np.linspace(0,50,5000)

sol = odeint(func = ode, y0 = initial_state, t = time_steps)
optimal_control = -R_inverse@B.T@P@sol.T

plt.plot(time_steps, optimal_control[0], label = "V_rc")
plt.plot(time_steps, optimal_control[1], label = "V_lc")
print(optimal_control[1][-1])
plt.ylim([-300,300])
plt.legend()
plt.show()

z_tilde = sol
z_desired = np.array([0, 0, V_nom, V_nom])
z = z_tilde + z_desired

Y_vals = z[:,0]
plt.plot(time_steps, Y_vals)
plt.title("Y")
plt.show()

theta_vals = z[:,1]
plt.plot(time_steps, theta_vals)
plt.title("Theta")
plt.show()

V_r_vals = z[:,2]
plt.plot(time_steps, V_r_vals)
plt.title("V_r")
plt.show()

V_l_vals = z[:,3]
plt.plot(time_steps, V_l_vals)
plt.title("V_l")
plt.show()
```
